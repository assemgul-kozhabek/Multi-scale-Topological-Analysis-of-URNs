```python
import networkx as nx
import csv
import matplotlib.pyplot as plt

# Read edge list from .csv file
edge_list_file = 'Surat.csv'
G = nx.Graph()

with open(edge_list_file, 'r') as f:
    reader = csv.reader(f)
    next(reader)  # Skip header row if present
    for row in reader:
        source = row[0]
        target = row[1]
        G.add_edge(source, target)

# Calculate the degree distribution
degree_counts = nx.degree_histogram(G)
degrees = range(len(degree_counts))

# Increase figure size
plt.figure(figsize=(7, 4.6))

# Create and display the histogram with grey bars
plt.bar(degrees, degree_counts, color='grey')
plt.title("Surat", fontsize=24)
plt.xlabel("Degree", fontsize=22)
plt.ylabel("Frequency", fontsize=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlim(0, 8)

plt.show()
```
