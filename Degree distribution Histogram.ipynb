```python
import networkx as nx
import matplotlib.pyplot as plt
import csv

# Create an example network
edge_list_file = 'Surat.csv'
G = nx.Graph()

# Load edge list from file
with open(edge_list_file, 'r') as f:
    reader = csv.reader(f)
    next(reader)  # Skip header row if present
    for row in reader:
        source = row[0]
        target = row[1]
        G.add_edge(source, target)
```

```python
# Calculate the degree distribution
degree_freq = nx.degree_histogram(G)

# Plot the degree distribution as a bar chart
plt.figure(figsize=(7, 5.5))  # Set figure size
plt.bar(range(len(degree_freq)), degree_freq, color='grey')
plt.xlabel('Degree', fontsize=22)
plt.ylabel('Frequency', fontsize=22)
plt.title('Surat', fontsize=24)
plt.xticks(range(len(degree_freq)), fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
